```python
import pandas as pd

df = pd.read_csv('02. Material de apoyo - Predicción del gasto en función del tiempo de navegación.csv')

df.head(), df.describe()
```

```python
import statsmodels.api as sm
X = sm.add_constant(df['tiempo_navegacion'])
Y = df['gasto']
modelo = sm.OLS(Y, X).fit()
print(modelo.summary())
```

```python
import numpy as np
pred = modelo.predict(X)
mae = np.mean(np.abs(Y - pred))
rmse = np.sqrt(np.mean((Y - pred)**2))
r2 = modelo.rsquared
print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
```
